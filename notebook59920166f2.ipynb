```python
import datetime as dt
import numpy as np
import pandas as pd
import os
import pickle

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv(os.path.join(dirname, filename))
df
```

```python
df["Release Date"] = pd.to_datetime(df["Release Date"])
```

```python
Dev = df.groupby('Developer').agg({'Title' : 'count', 'User Score' : 'mean'}).reset_index().sort_values(by = 'Title', ascending=False)
Dev['Cumulative']=(Dev['Title']/Dev['Title'].sum()).cumsum()
Dev=Dev.head(100)

fig=go.Figure()
fig.update_layout(yaxis2=dict(overlaying='y',side='right'))

bar_color = 'yellow' 
scatter_color = 'white'  

trace_1=go.Bar(x=Dev['Developer'],y=Dev['Title'],name='nombres de jeux sortis', marker_color = bar_color)
trace_2=go.Scatter(x=Dev['Developer'],y=Dev['User Score'],yaxis='y2',mode='lines+markers',name='Moyen des notes des joueurs', marker_color = scatter_color)
fig.add_traces([trace_1,trace_2])

fig.update_layout(height=600,template='plotly_dark')
fig.update_traces(selector=dict(type='bar'))
fig.update_traces(selector=dict(type='scatter'), marker=dict(size=Dev['User Score']))

fig.update_layout(title=dict(text='Jeux par Developpeurs'))
#fig.update_layout(yaxis2=dict(tickformat='.0%'))
fig.update_xaxes(tickvals=Dev['Developer'],tickangle=45,tickfont_size=8)
```

```python
Dev=df.groupby('Publisher').agg({'Title':'count','User Score':'mean'}).reset_index().sort_values(by='Title',ascending=False)
Dev['Cumulative']=(Dev['Title']/Dev['Title'].sum()).cumsum()
Dev=Dev.head(100)

fig=go.Figure()
fig.update_layout(yaxis2=dict(overlaying='y',side='right'))

bar_color = 'yellow' 
scatter_color = 'white' 

trace_1=go.Bar(x=Dev['Publisher'],y=Dev['Title'],name='Nombres jeux sortis par editeur', marker_color = bar_color)
trace_2=go.Scatter(x=Dev['Publisher'],y=Dev['User Score'],yaxis='y2',mode='lines+markers',name='Moyenne score ',marker_color = scatter_color)
fig.add_traces([trace_1,trace_2])

fig.update_layout(height=600,template='plotly_dark')
fig.update_traces(selector=dict(type='bar'))
fig.update_traces(selector=dict(type='scatter'), marker=dict(size=Dev['User Score']))

fig.update_layout(title=dict(text='Nombres jeux sortis par editeur'))
#fig.update_layout(yaxis2=dict(tickformat='.0%'))
fig.update_xaxes(tickvals=Dev['Publisher'],tickangle=45,tickfont_size=8)
```

```python
fig=px.box(df,x='User Score',y='User Ratings Count')
fig.update_yaxes(type='log')
fig.update_layout(template='plotly_dark')
fig.update_layout(title=dict(text='Fréquence de Distribution des évaluations données par les utilisateurs en fonction de la note des utilisateurs'))
```

```python
Dev=df.groupby(df['Release Date'].dt.to_period('Q')).agg({'Title':'count','User Score':'mean'}).reset_index()
Dev=Dev.iloc[:-1,:]
Dev['Release Date']=Dev['Release Date'].astype(str)

fig=go.Figure()
fig.update_layout(yaxis2=dict(overlaying='y',side='right'))

trace_1=go.Scatter(name='Number OF Games Released',x=Dev['Release Date'],y=Dev['Title'],mode='lines+markers')
trace_2=go.Scatter(name='AVG User Score',x=Dev['Release Date'],y=Dev['User Score'],yaxis='y2',mode='lines+markers')
fig.add_traces([trace_1,trace_2])

fig.update_layout(height=600,template='plotly_dark')
fig.update_layout(title=dict(text='Nombre de jeux sortis et moyenne des joueurs'))
```

```python
Dev=df['Product Rating'].value_counts()

fig=px.pie(labels=Dev.index,values=Dev.values,hover_name=Dev.index)
fig.update_layout(template='plotly_dark')
fig.update_layout(title=dict(text='Contenu video par catégorie de produit'))
```

```python
Dev=df.groupby(['Product Rating',df['Release Date'].dt.to_period('Y')])['Title'].count().sort_index(level=1).reset_index()
Dev['Percentage']=Dev.groupby('Release Date')['Title'].transform(lambda x:x/x.sum())*100
Dev['Percentage']=Dev['Percentage'].apply(lambda x:f'{x:.0f} %')
Dev['Release Date']=Dev['Release Date'].astype(str)
Dev=Dev[Dev['Product Rating'].isin(df['Product Rating'].value_counts().index[:4].tolist())]

fig=px.bar(Dev,x='Release Date',y='Title',color='Product Rating',text='Percentage')
fig.update_layout(height=600,template='plotly_dark')
fig.update_layout(title=dict(text='Création de contenu par catégorie jeux sur un an'))
fig.update_yaxes(title=dict(text='Number Of Games Released'))
```
